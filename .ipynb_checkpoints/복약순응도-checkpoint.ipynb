```python
#단일처방
#성분검색
def Search_Pharm(df):
    string_1 = str(input('약품일반성분명코드 입력하시오'))
    Visit = df.loc[df['약품일반성분명코드'] == string_1]
    return Visit

#방문횟수 카운트
def Visit_count(df):
    count = pd.DataFrame(df.groupby(['약품일반성분명코드'])['가입자일련번호'].value_counts()).rename(columns = {'가입자일련번호' : '방문횟수'})
    count = count.reset_index()
    #방문횟수가 3번이상인 사람부터 연산
    count = count[count['방문횟수'] > 2]
    count = count.drop(['약품일반성분명코드'], axis = 1)
    final = pd.merge(count, df, on = '가입자일련번호', how = 'left')
    final = final.drop_duplicates(['가입자일련번호', '요양개시일자', '약품일반성분명코드'])
    final = final.sort_values(['가입자일련번호', '요양개시일자'])
    return final


#처방일 간격 계산
def Calculate(df):
    # for문 _ 가입자별 요양 개시일자 창
    Date_dif_1 = df['요양개시일자'][1:].reset_index().drop(['index'], axis = 1).rename(columns = {'요양개시일자' : '처방일간격'})
    Date_dif_2 = df['요양개시일자'][0:].reset_index().drop(['index'], axis = 1).rename(columns = {'요양개시일자' : '처방일간격'})
    Date_dif = Date_dif_1 - Date_dif_2[:-1]
    final = pd.concat([df, Date_dif], axis = 1)
    #final = final.reset_index()
    #final = final.dropna(axis = 0)
    #final = final.drop(['index'], axis = 1)
    return final

#날짜 type을 int로 변경(days 삭제)    
def Date_Number(df):
    change = df['처방일간격'].astype(str).str.slice(0 , -24).astype(int)
    data = df.drop(['처방일간격'], axis = 1)
    final = pd.concat([data, change], axis = 1)
    #마이너스 제거
    final_del = final[final['처방일간격'] < 0].index
    final = final.drop(final_del)
    final = final.reset_index()
    final = final.drop(['index'], axis = 1)
    return final

## Calculate + Date_Number
#def Calculate(df):
    

#복약순응도 추정
def Medication(df):
    df['복약순응도'] = (df['총투여일수']/df['처방일간격'])*100
    df['복약순응도'].loc[df['복약순응도'] > 100] = 100
    mean = df.groupby('가입자일련번호')['복약순응도'].mean()
    #환자별 평균복약순응도 산출
    df['평균복약순응도'] = df['가입자일련번호'].map(mean)
    final = df
    return final

#복약순응도 결과정리(처방건수기준)
def display_1(df):
    fig = plt.figure(figsize = [12, 6])
    fig = plt.title('복약순응도 시각자료(처방건수)')
    fig = sns.distplot(df['복약순응도'], color='blue', label = '복약순응도')
    final = df['복약순응도'].describe([.10, .20, .30, .40, .60, .70, .80, .90])
    final = final.astype(int)
    return final

#복약순응도 결과정리(환자평균)
def display_2(df):
    mean = df.drop_duplicates('가입자일련번호')
    fig = plt.figure(figsize = [12, 6])
    fig = plt.title('복약순응도 시각자료(환자평균)')
    fig = sns.distplot(mean['평균복약순응도'], color='blue', label = '복약순응도')
    final = mean['평균복약순응도'].describe([.10, .20, .30, .40, .60, .70, .80, .90])
    final = final.astype(int)
    return fig, final
```

```python
#병용처방 2가지
def Search_2(df):
    str_1, str_2 = str(input('약품일반성분명코드 입력하시오')).split()
    pharm_data = df[df['약품일반성분명코드'].isin([str_1, str_2])]
    pres_data = pd.DataFrame(pharm_data['처방내역일련번호'].value_counts() > 1).reset_index()
    pres_data = pres_data.rename(columns = {'처방내역일련번호' : '불리언', 'index' : '처방내역일련번호'})
    data = pd.merge(pharm_data, pres_data, on= '처방내역일련번호',how = 'left')
    data = data.loc[data['불리언'] == True]
    data = data.sort_values(['요양개시일자'])
    data = data.reset_index()
    data = data.drop(['index'], axis = 1)
    return data

#병용처방 3가지
def Search_3(df):
    str_1, str_2, str_3 = str(input('약품일반성분명코드 입력하시오')).split()
    pharm_data = df[df['약품일반성분명코드'].isin([str_1, str_2, str_3])]
    pres_data = pd.DataFrame(pharm_data['처방내역일련번호'].value_counts() > 2).reset_index()
    pres_data = pres_data.rename(columns = {'처방내역일련번호' : '불리언', 'index' : '처방내역일련번호'})
    data = pd.merge(pharm_data, pres_data, on= '처방내역일련번호',how = 'left')
    data = data.loc[data['불리언'] == True]
    data = data.sort_values(['요양개시일자'])
    data = data.reset_index()
    data = data.drop(['index'], axis = 1)
    return data

#병용처방 4가지
def Search_4(df):
    str_1, str_2, str_3, str_4 = str(input('약품일반성분명코드 입력하시오')).split()
    pharm_data = df[df['약품일반성분명코드'].isin([str_1, str_2, str_3, str_4])]
    pres_data = pd.DataFrame(pharm_data['처방내역일련번호'].value_counts() > 3).reset_index()
    pres_data = pres_data.rename(columns = {'처방내역일련번호' : '불리언', 'index' : '처방내역일련번호'})
    data = pd.merge(pharm_data, pres_data, on= '처방내역일련번호',how = 'left')
    data = data.loc[data['불리언'] == True]
    data = data.sort_values(['요양개시일자'])
    data = data.reset_index()
    data = data.drop(['index'], axis = 1)
    return data

#병용처방 5가지
def Search_5(df):
    str_1, str_2, str_3, str_4, str_5 = str(input('약품일반성분명코드 입력하시오')).split()
    pharm_data = df[df['약품일반성분명코드'].isin([str_1, str_2, str_3, str_4, str_5])]
    pres_data = pd.DataFrame(pharm_data['처방내역일련번호'].value_counts() > 4).reset_index()
    pres_data = pres_data.rename(columns = {'처방내역일련번호' : '불리언', 'index' : '처방내역일련번호'})
    data = pd.merge(pharm_data, pres_data, on= '처방내역일련번호',how = 'left')
    data = data.loc[data['불리언'] == True]
    data = data.sort_values(['요양개시일자'])
    data = data.reset_index()
    data = data.drop(['index'], axis = 1)
    return data

#병용처방 6가지
def Search_6(df):
    str_1, str_2, str_3, str_4, str_5, str_6 = str(input('약품일반성분명코드 입력하시오')).split()
    pharm_data = df[df['약품일반성분명코드'].isin([str_1, str_2, str_3, str_4, str_5, str_6])]
    pres_data = pd.DataFrame(pharm_data['처방내역일련번호'].value_counts() > 5).reset_index()
    pres_data = pres_data.rename(columns = {'처방내역일련번호' : '불리언', 'index' : '처방내역일련번호'})
    data = pd.merge(pharm_data, pres_data, on= '처방내역일련번호',how = 'left')
    data = data.loc[data['불리언'] == True]
    data = data.sort_values(['요양개시일자'])
    data = data.reset_index()
    data = data.drop(['index'], axis = 1)
    return data
```
